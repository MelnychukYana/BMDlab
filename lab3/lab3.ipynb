```python
import seaborn as sns
%matplotlib inline
```

Збереження даних

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

Візуалізація даних

1.displot. Гістограма та графік щільності росподілу

```python
#displot - показує розподіл одновимірного набору спостережень
#kde - за замовчуванням True, якщо kde=False - то на графіку буде лише гістограма без кривої розподілу
#bins - ширина "кроку" для гістограми
sns.distplot(flights['passengers'], bins=20)
```

```python
sns.jointplot(x='passengers', y='year', data = flights, kind="hex", color="y", height = 5)
```

```python
sns.pairplot(flights)
```

```python
sns.pairplot(flights, hue = 'month', palette = "Set2")
```

Графіки для аналізу категоріальних даних:

1.barplot – дозволяють отримувати сукупні дані по кількісним даним в залежності від категоріальної змінної.

```python
sns.barplot(x = 'year', y = 'passengers', data = flights)
```

```python
import numpy as np
sns.barplot(x = 'month', y = 'passengers', data = flights, estimator =np.median, )
```

2.boxplots - використовується для візуалізації розподілу категоріальних даних за певною кількісною змінною.

```python
#boxplots - використовується для візуалізації розподілу категоріальних даних за певною кількісною змінною
#orient - орієнтація ділянки (вертикальна 'v' або горизонтальна 'v')

sns.boxplot(x = 'passengers', y = 'month', data = flights, orient='h', palette = 'rainbow')
```

3.stripplot – представляє собою діаграму розсіювання, де одна за змінних категоріальна.

```python
sns.stripplot(x = 'month', y = 'passengers', data = flights, size = 6, palette = 'coolwarm' )
```

```python
sns.heatmap(flights.corr())
```

```python
#annot - якщо annot = True записуються значення даних у кожну клітинку (annot – додавання коефіцієнтів кореляції на діаграму)
```

```python
sns.heatmap(flights.corr(), cmap = 'Set1', annot = True)
```

Графіки для відображення даних у вигляді сітки

```python
import matplotlib.pyplot as plt
b = sns.FacetGrid(flights, row = 'month', row_order = ['July','May'])
b = b.map(plt.hist, 'passengers', color = 'y')
```

```python
c = sns.FacetGrid(flights,  row = "month")
c = c.map(plt.hist, "passengers")
```
