1. Основи побудови графіків

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
y
```

2. Побудова графіків2.1 Функціональний метод

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title("y = x^2")
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'palegoldenrod')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'lightseagreen')
```

2.2 Об'єктно-орієнтований метод

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y)
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('y=x^2')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.1,0.5,0.4,0.2])

axes1.plot(x,y)
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('y=x^2')

axes2.plot(y,x)
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_title('y=x^2')
```

Створення декількох графіків, використовуючи subplots

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y,'r')
axes[0].set_xlabel("X label")
axes[0].set_ylabel("Y label")
axes[0].set_title("Graph 1")

axes[1].plot(y,x,'b')
axes[1].set_xlabel("Y label")
axes[1].set_ylabel("X label")
axes[1].set_title("Graph 2")
```

```python
fig, axes = plt.subplots(nrows=5, ncols=5)
```

3.Редагування графіків.
Додавання легенди на графік.

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x**3, label='x^3')
axes.plot(x, x**(1/2), label='sqrt(x)')
axes.legend()
```

Змінення різних характеристик графіку.

```python
fig = plt.figure(figsize=(8,4), dpi=100)

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2, label="x**2", color="g")
ax.plot(x, x**3, label="x**3", color="y")
ax.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**2, lw = 3, ls = '--')
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**2, marker = 'o', markersize = 10, markerfacecolor = 'g')
```
